```python
import os
import random

# 1. Point to your Google Drive folder
# Update this string to match the exact folder name where you saved the files in Drive
data_dir = "/content/drive/MyDrive/JSON"
all_files = [f for f in os.listdir(data_dir) if f.endswith('.json')]

# Randomly sample 1000 patients, or fewer if not enough are available
random.seed(42)
sampled_files = random.sample(all_files, min(len(all_files), 1000))
```

```python
# 2. Extract specific resources (e.g., Observations)
observations_data = []
for file_name in sampled_files:
    file_path = os.path.join(data_dir, file_name)
    with open(file_path, 'r') as f:
        fhir_bundle = json.load(f)

        # Check if 'entry' exists in the bundle
        if 'entry' in fhir_bundle:
            for entry in fhir_bundle['entry']:
                resource = entry.get('resource', {})

                # Filter for the specific resource type
                if resource.get('resourceType') == 'Observation':
                    observations_data.append(resource)
```

```python
# 3. Flatten the nested JSON into a clean, tabular format
df_observations = pd.json_normalize(observations_data)

# 4. Save the CSV directly to your Drive so it is permanently stored
output_path = "/content/drive/MyDrive/flattened_observations.csv"
df_observations.to_csv(output_path, index=False)

print(f"Flattened {len(df_observations)} observations into tabular format. Saved to {output_path}.")
```

```python
import pandas as pd
import numpy as np
import ast
```

```python
df = pd.read_csv('flattened_observations.csv')

print("Original Shape:", df.shape)
```

```python
df.info()
```

```python
# Remove duplicates
df = df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)
```

```python
df.head()
```

```python
# Convert datetime
df["effectiveDateTime"] = pd.to_datetime(
    df["effectiveDateTime"],
    errors="coerce"
)
```

```python
df.info()
```

```python
# Keep only final observations
if "status" in df.columns:
    df = df[df["status"] == "final"]
```

```python
df.info()
```

```python
# Create date features
df["observation_date"] = df["effectiveDateTime"].dt.date
df["year"] = df["effectiveDateTime"].dt.year
df["month"] = df["effectiveDateTime"].dt.month
df["month_name"] = df["effectiveDateTime"].dt.month_name()

# Convert numeric observation values
df["valueQuantity.value"] = pd.to_numeric(
    df["valueQuantity.value"],
    errors="coerce"
)
```

```python
df.info()
```

```python
def extract_code_info(code_str):

    try:
        code_list = ast.literal_eval(code_str)

        if isinstance(code_list, list) and len(code_list) > 0:

            first_item = code_list[0]

            return pd.Series({
                "loinc_code": first_item.get("code"),
                "observation_name": first_item.get("display")
            })

    except:
        pass

    return pd.Series({
        "loinc_code": np.nan,
        "observation_name": np.nan
    })


code_details = df["code.coding"].apply(extract_code_info)

df = pd.concat([df, code_details], axis=1)
```

```python
df.head()
```

```python
def extract_bp_values(component_str):

    systolic = np.nan
    diastolic = np.nan

    try:
        components = ast.literal_eval(component_str)

        for item in components:

            coding = item.get("code", {}).get("coding", [])

            if len(coding) > 0:

                code = coding[0].get("code")

                value = item.get(
                    "valueQuantity", {}
                ).get("value")

                # Systolic BP
                if code == "8480-6":
                    systolic = value

                # Diastolic BP
                elif code == "8462-4":
                    diastolic = value

    except:
        pass

    return pd.Series({
        "systolic_bp": systolic,
        "diastolic_bp": diastolic
    })


bp_values = df["component"].apply(extract_bp_values)

df = pd.concat([df, bp_values], axis=1)
```

```python
df.head()
```

```python
df["bmi_category"] = None

df.loc[
    (df["observation_name"] == "Body Mass Index") &
    (df["valueQuantity.value"] >= 30),
    "bmi_category"
] = "Obese"

df.loc[
    (df["observation_name"] == "Body Mass Index") &
    (df["valueQuantity.value"] >= 25) &
    (df["valueQuantity.value"] < 30),
    "bmi_category"
] = "Overweight"

df.loc[
    (df["observation_name"] == "Body Mass Index") &
    (df["valueQuantity.value"] < 25),
    "bmi_category"
] = "Normal"
```

```python
df.head()
```

```python
df["high_bp_flag"] = np.where(

    (df["systolic_bp"] >= 140) |
    (df["diastolic_bp"] >= 90),

    1,
    0
)
```

```python
df.head()
```

```python
df["diabetes_risk"] = None

df.loc[
    (
        df["observation_name"] ==
        "Hemoglobin A1c/Hemoglobin.total in Blood"
    ) &
    (df["valueQuantity.value"] >= 6.5),
    "diabetes_risk"
] = "Diabetic"

df.loc[
    (
        df["observation_name"] ==
        "Hemoglobin A1c/Hemoglobin.total in Blood"
    ) &
    (df["valueQuantity.value"] >= 5.7) &
    (df["valueQuantity.value"] < 6.5),
    "diabetes_risk"
] = "Prediabetic"

df.loc[
    (
        df["observation_name"] ==
        "Hemoglobin A1c/Hemoglobin.total in Blood"
    ) &
    (df["valueQuantity.value"] < 5.7),
    "diabetes_risk"
] = "Normal"
```

```python
df.head()
```

```python
df.columns = [

    col.replace(".", "_")
       .replace(" ", "_")
       .lower()

    for col in df.columns
]
```

```python
df.columns
```

```python
dashboard_df = df[
    [
        'id', 'status', 'effectivedatetime', 'resourcetype', 'code_coding',
       'subject_reference', 'encounter_reference', 'valuequantity_value',
       'valuequantity_unit', 'valuequantity_system', 'valuequantity_code',
       'component', 'valuecodeableconcept_coding', 'observation_date', 'year',
       'month', 'month_name', 'loinc_code', 'observation_name', 'systolic_bp',
       'diastolic_bp', 'bmi_category', 'high_bp_flag', 'diabetes_risk'
    ]
]


# Remove empty observation rows
dashboard_df = dashboard_df.dropna(
    subset=["observation_name"]
)
```

```python
output_file = "looker_ready_healthcare_dataset.csv"

dashboard_df.to_csv(
    output_file,
    index=False
)

print("\nCleaned Dataset Shape:", dashboard_df.shape)

print("\nCSV Exported Successfully:")
print(output_file)
```
